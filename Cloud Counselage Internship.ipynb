Q. Average family income of students?

```python
import re
import numpy as np

def convert_income_to_numeric(income_str):
    # Find all the numbers in the string
    numbers = list(map(int, re.findall(r'\d+', income_str)))

    if len(numbers) == 2:
        # Calculate the mid-point of the range
        return np.mean(numbers)
    elif len(numbers) == 1:
        # Return the single number
        return numbers[0]
    else:
        # Handle cases where the format is unexpected
        return np.nan

# Check if the 'Family Income' column exists
if 'Family Income' in df.columns:
    # Apply the conversion function to the 'Family Income' column
    df['Family Income Numeric'] = df['Family Income'].apply(convert_income_to_numeric)

    # Calculate the average family income
    average_family_income = df['Family Income Numeric'].mean()
    print(f'The average family income of the students is: {average_family_income:.2f} Lakh')
else:
    print("The 'Family Income' column does not exist in the dataset.")
```

Q. Those who have high experience in Language and high CGPA and those who have high expected salary? (Avg)

```python
import pandas as pd

# Sample data
data = {
    'CGPA': ['8.5', '7.0', '9.0', '6.5'],
    'Experience with python (Months)': ['24', '12', '36', '48'],
    'Expected salary (Lac)': ['5 Lac', '3 Lac', '8 Lac', '6 Lac']
}
df = pd.DataFrame(data)

# Convert columns to appropriate data types
df['CGPA'] = pd.to_numeric(df['CGPA'], errors='coerce')
df['Experience with python (Months)'] = pd.to_numeric(df['Experience with python (Months)'], errors='coerce')

# Ensure 'Expected salary (Lac)' is treated as a string for cleaning
df['Expected salary (Lac)'] = df['Expected salary (Lac)'].astype(str)

# Clean and convert 'Expected salary (Lac)' to numeric
df['Expected Salary'] = pd.to_numeric(
    df['Expected salary (Lac)']
    .str.replace('Lac', '')
    .str.replace(',', '')
    .str.strip(),
    errors='coerce'
)

# Define thresholds for high CGPA and high experience
high_cgpa_threshold = 8.0  # Example threshold for high CGPA
high_experience_threshold = 24  # Example threshold for months of experience

# Filter students with high CGPA and high experience
high_cgpa_experience_students = df[
    (df['CGPA'] >= high_cgpa_threshold) &
    (df['Experience with python (Months)'] >= high_experience_threshold)
]

# Calculate the average expected salary for these students
average_expected_salary = high_cgpa_experience_students['Expected Salary'].mean()

# Print the result
print(f'Average expected salary for students with high CGPA and high experience: {average_expected_salary:.2f} Lac')
```
